```python
import numpy as np
import matplotlib.pyplot as plt

# サンプルデータ生成
fs = 1000  # サンプリング周波数
t = np.arange(0, 1, 1/fs)  # 1秒間の時間軸
f1 = 10  # 周波数10Hzの成分
f2 = 50  # 周波数50Hzの成分
signal = np.sin(2 * np.pi * f1 * t) + 0.5 * np.sin(2 * np.pi * f2 * t)

# フーリエ変換
fft_result = np.fft.fft(signal)
fft_freq = np.fft.fftfreq(len(t), 1/fs)

# ローパスフィルタ
cutoff_frequency = 30  # カットオフ周波数
fft_result_filtered = fft_result.copy()
fft_result_filtered[np.abs(fft_freq) > cutoff_frequency] = 0

# 逆フーリエ変換
filtered_signal = np.fft.ifft(fft_result_filtered)

# 画像を4分割してプロット
plt.figure(figsize=(10, 6))


# 左上: 元の信号
plt.subplot(2, 2, 1)
plt.plot(t, signal)
plt.title('Original Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# 左下: 元の信号のフーリエ変換
plt.subplot(2, 2, 3)
plt.plot(fft_freq, np.abs(fft_result))
plt.title('Fourier Transform')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')

# 右下: フィルタをかけた結果（周波数領域）
plt.subplot(2, 2, 4)
plt.plot(fft_freq, np.abs(fft_result_filtered))
plt.title('Filtered Signal (Frequency Domain)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')

# 右上: 逆フーリエ変換を行った信号
plt.subplot(2, 2, 2)
plt.plot(t, np.real(filtered_signal))  # np.real()を使用して実部を取得
plt.title('Signal after Inverse Fourier Transform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# サンプルデータ生成
fs = 1000  # サンプリング周波数
t = np.arange(0, 1, 1/fs)  # 1秒間の時間軸
f1 = 10  # 周波数10Hzの成分
f2 = 50  # 周波数50Hzの成分
signal = np.sin(2 * np.pi * f1 * t) + 0.5 * np.sin(2 * np.pi * f2 * t)

# フーリエ変換
fft_result = np.fft.fft(signal)
fft_freq = np.fft.fftfreq(len(t), 1/fs)

# ローパスフィルタ
cutoff_frequency = 30  # カットオフ周波数
fft_result_filtered = fft_result.copy()
fft_result_filtered[np.abs(fft_freq) > cutoff_frequency] = 0

# 逆フーリエ変換
filtered_signal = np.fft.ifft(fft_result_filtered)

# 画像の4分割
plt.figure(figsize=(10, 6))

# 左上: 元の信号
plt.subplot(2, 2, 1)
plt.plot(t, signal)
plt.title('Original Signal')

# 左下: 元の信号のフーリエ変換
plt.subplot(2, 2, 3)
plt.plot(fft_freq, np.abs(fft_result))
plt.title('Fourier Transform')

# 右下: フィルタをかけた結果
plt.subplot(2, 2, 4)
plt.plot(t, filtered_signal)
plt.title('Filtered Signal (Low-pass)')

# 右上: 逆フーリエ変換を行った信号
plt.subplot(2, 2, 2)
plt.plot(t, np.real(np.fft.ifft(fft_result)))
plt.title('Inverse Fourier Transform')

plt.tight_layout()
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# グラフの作成
G = nx.Graph()

# ノードとエッジの追加
nodes = [('Paper1', {'Type': 'Physics'}),
         ('Paper2', {'Type': 'Physics'}),
         ('Paper3', {'Type': 'Physics'}),
         ('Paper4', {'Type': 'Mathematics'}),
         ('Paper5', {'Type': 'Mathematics'}),
         ('Paper6', {'Type': 'Mathematics'}),
         ('Paper7', {'Type': 'Chemistry'}),
         ('Paper8', {'Type': 'Chemistry'}),
         ('Paper9', {'Type': 'Chemistry'}),
         ('Paper10', {'Type': 'Physics'})]

edges = [('Paper1', 'Paper2'), ('Paper1', 'Paper3'), ('Paper2', 'Paper4'),
         ('Paper2', 'Paper5'), ('Paper3', 'Paper6'), ('Paper3', 'Paper7'),
         ('Paper4', 'Paper8'), ('Paper5', 'Paper9'), ('Paper6', 'Paper10')]

G.add_nodes_from(nodes)
G.add_edges_from(edges)

# グラフのプロット
pos = nx.spring_layout(G)
node_colors = [0 if G.nodes[n]['Type'] == 'Physics' else 1 if G.nodes[n]['Type'] == 'Mathematics' else 2 for n in G.nodes]

# カラーマップの設定
cmap = plt.cm.get_cmap('Set1', 3)

# グラフの描画
nx.draw(G, pos, with_labels=True, node_color=node_colors, cmap=cmap, font_weight='bold')

# カラーバーの追加
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=0, vmax=2))
sm.set_array([])
cbar = plt.colorbar(sm, ticks=[0, 1, 2], orientation='vertical')
cbar.set_label('Paper Type')

# グラフの表示
plt.show()
```
