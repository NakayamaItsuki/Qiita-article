```python
import torch
import matplotlib.pyplot as plt

# データの定義
x_data = torch.tensor([1, 3, 4, 7, 10])
y_data = torch.tensor([2, 5, 9, 3, 1])

# Matplotlibを使用してプロット
plt.plot(x_data.numpy(), y_data.numpy(), marker='o', linestyle='-')

# 各点の下にx軸の値を表示
for x, y in zip(x_data, y_data):
    plt.text(x.numpy(), -0.5, f'{x.item()}', ha='center')

# グラフにラベルを追加
plt.title('Plot with X-axis Values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# グリッドを表示
plt.grid(True)

# プロットを表示
plt.show()
```

```python
import torch
import pygsp

# ここで適当なグラフを生成するか、データセットから読み込んでください
# 以下は適当な例です（適宜変更してください）
G = pygsp.graphs.Logo()

# グラフ信号の生成
# ここでは適当な信号を生成していますが、実際のデータに置き換えてください
signal = torch.randn(G.N)

# PyGSPのGraphオブジェクトを作成
graph = pygsp.graphs.Graph(W=G.W, coords=G.coords)

# グラフフーリエ変換を行う
fourier_basis = pygsp.filters.compute_fourier_basis(graph)
coefficients = pygsp.graphs.fourier(signal, fourier_basis)

# グラフフーリエ逆変換を行う（信号の復元）
reconstructed_signal = pygsp.graphs.inverse_fourier(coefficients, fourier_basis)

# 結果の表示
print("元の信号:", signal)
print("フーリエ係数:", coefficients)
print("復元された信号:", reconstructed_signal)
```
